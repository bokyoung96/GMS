```python
merged_df = pd.read_csv("ticker_final.csv")
```

```python
merged_df_1 = merged_df[merged_df['Date'] == '2019-12-31']
ticker_list = merged_df_1['ticker'].tolist()
```

```python
#### download_data 함수를 통해 종목의 재무정보 다운받음 

def download_data(ticker):
    try:
        fd =['TR.CompanyMarketCap(SDate=2019-12-01,EDate=2020-03-31,Frq=M)',
        'TR.PE(SDate=2019-12-01,EDate=2020-03-31,Frq=M)',
         'TR.PriceToSalesPerShare(SDate=2019-12-01,EDate=2020-03-31,Frq=M)',
        'TR.EVToEBITDA(SDate=2019-12-01,EDate=2020-03-31,Frq=M)',
        'TR.PricetoCFPerShare(SDate=2019-12-01,EDate=2020-03-31,Frq=M)',
         'TR.PriceToBVPerShare(SDate=2019-12-01,EDate=2020-03-31,Frq=M)',
          'TR.DividendYield(SDate=2019-12-01,EDate=2020-03-31,Frq=M)',
          'TR.TotalReturn1Mo(SDate=2019-12-01,EDate=2020-03-31,Frq=M)',
         'TR.PriceClose(SDate=2019-12-01,EDate=2020-03-31,Frq=M).date']

        data, err=ek.get_data(ticker ,fields=fd)
    
        return data
    
    except Exception as e:
        print(f"Error occurred: {e}")
        pass
```

```python
results = []

for ticker in ticker_list:
    data = download_data(ticker)
    if data is not None:
        results.append(data)
        print(ticker)
```

```python
final_datafram_2 = pd.concat(results, ignore_index=True)
df_final = final_datafram_2[['Instrument', 'Company Market Cap','P/E (Daily Time Series Ratio)', 'Price To Sales Per Share (Daily Time Series Ratio)','Enterprise Value To EBITDA (Daily Time Series Ratio)',
                                'Price To Cash Flow Per Share (Daily Time Series Ratio)','Price To Book Value Per Share (Daily Time Series Ratio)','Dividend yield', '1 Month Total Return', 'Date']]
```

```python
def date_change(date):
    date = str(date)
    if '2019-12' in date:
        return 'M1'
    elif '2020-01' in date :
        return 'M2'
    elif '2020-02' in date :
        return 'M3'
    elif '2020-03' in date  :
        return 'M4'
    else: 
        return None
    
    
df_final['Date_period'] = df_final['Date'].apply(date_change)
df_final.to_csv("final_data.csv")
```
